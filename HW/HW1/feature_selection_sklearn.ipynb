```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
```

```python
train_path = './HW1.train.csv'  # path to training data
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
df = pd.read_csv(train_path, index_col=0)
df.head()
```

```python
state_col = df.iloc[:,:40]
data_col = df.iloc[:,40:-1]
```

```python
data_col.head()
```

```python
selector = SelectKBest(f_regression, k=15).fit(data_col, df.iloc[:,-1])
feats_col =pd.concat([data_col.iloc[:,selector.get_support()]], axis=1)
feats_selected = [df.columns.get_loc(col) for col in feats_col.columns]
feats_selected
```

```python
df.iloc[:,feats_selected].columns
```

```python
p_values = selector.pvalues_
p_values.sort()
p_values
```

```python
scores = selector.scores_
scores.sort()
scores
```
